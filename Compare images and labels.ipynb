```python
import os
import glob
```

```python
image_dir = "images/PhM_SFDC"
annotation_dir = 'annotations/philip_morris_full'
```

```python
# get image list
images = glob.glob1(image_dir, '*.jpg')
len(images)
```

```python
# get annotation list
annot = glob.glob1(annotation_dir, "UA*.xml")
len(annot)
```

```python
for i in annot:
    if os.path.splitext(i)[0] + '.jpg' not in images:
        print("Image not found. %s" % os.path.splitext(i)[0])
        os.remove(os.path.join(annotation_dir, i))
```
