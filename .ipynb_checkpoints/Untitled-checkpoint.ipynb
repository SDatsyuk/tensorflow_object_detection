```python
import numpy as np
import os
import cv2
```

```python
image_path = 'test_images/image1.jpg'
```

```python
image = cv2.imread(image_path)
im_height, im_width, _ = image.shape
```

```python
results = np.load('res.npy')
results.item()
```

```python
data = results.item()
list(zip(data['detection_boxes'], data['detection_classes'], data['detection_scores']))
```

```python
detections = [(box, score, label) for box, label, score in zip(data['detection_boxes'], data['detection_classes'], data['detection_scores']) if score > 50]
```

```python
detections
```

```python
for (box, score, label) in detections:
    ymin, xmin, ymax, xmax = box.tolist()
    (left, right, top, bottom) = [int(i) for i in (xmin * im_width, xmax * im_width, ymin * im_height, ymax * im_height)]
    cv2.imshow('ss', image[top:bottom, left:right])
    cv2.waitKey(0)
```
